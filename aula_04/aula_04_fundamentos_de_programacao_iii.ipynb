## Funções

Podemos definir valores padrões para os parâmetros de uma função, o que significa que, se um argumento não for fornecido quando a função for chamada, ele usará o valor padrão definido em vez disso.

```python
def minha_funcao(parametro_a,parametro_b,parametro_c):
    comprimento = parametro_a
    altura=parametro_b
    espessura = parametro_c
    volume = comprimento*altura*espessura
    print(f'O volume do paralelepipedo é {volume}')
```

```python
minha_funcao(2,30,1)
```

```python
minha_funcao(30,40,60)
```

```python
def pai():
    print('oi eu sou a parte do codigo pai')

    def filha_1():
        print('oi, eu sou a responsividade da função filha_1')
    def filha_2():
        print('oi eu sou a filha_2')

    filha_1()
    filha_2()
```

```python
pai()
```

```python
def calcular(simbolo):
    def soma(a,b):
        return a+b
    def subtracao(a,b):
        return a-b
    
    if simbolo == "+":
        return soma
    elif simbolo =="-":
        return subtracao
    else:
        print('não impletado')
        return None
```

```python
calcular("-")(1,2)
```

```python
def calcular(simbolo):
    def soma(a,b):
        return a+b
    def subtracao(a,b):
        return a-b
    def multiplicacao(a,b):
        return a*b
    def divisao(a,b):
        if b == 0:
            raise Exception('o segundo parametro não pode ser zero') 
        return a/b
    
    match simbolo:
        case "+":
            return soma
        case "-":
            return subtracao
        case "*":
            return multiplicacao
        case "/":
            return divisao
        case _:
            print("simbolo invalido")
```

```python
calcular("*")(4,3)
```

## Map


map é uma função Built-in que recebe uma função e uma estrutura de dados como argumentos, e aplica a função a cada elemento da estrutura, retornando um dado do tipo map, que mapeou cada item da estrutura anterior. Para tornar funcional, só criar um cast da variável armazenada pelo map, podendo ser list(), tuple() ou set().


```python
frutas = ['banana','pera','uva']

quantidade_letras = map(len,frutas)
```

```python
print(quantidade_letras)
```

```python
lista_len = list(quantidade_letras)
print(lista_len)
```

```python
tamanho_nome = map(len,('joão','maria','jose'))
print(list(tamanho_nome))
```

```python
quadrados = map(quadrado,{1,2,3})
print(list(quadrados))
```

crie uma função chamada substitui_a_por_x que vai substituir a por x

depois crie uma lista  a partir da lista  nomes: joão, maria, mariana, ,juliana, pedro

crie a nova lista com todos os valores na lista com as vogais a trocadas.

```python
def substitui_a_por_x(palavra):
    return palavra.replace('a','x').replace('á','x').replace('ã','x')

lista_nomes = ['joão','maria','mariana']

resultado = map(substitui_a_por_x,lista_nomes)

nova_lista = list(resultado)

print(nova_lista)
```
